# Word Guessing Game

Welcome to the Word Guessing Game! First let's import the files that will define our list of words in a function, so we can call that later in-game. For this game, the file "words_alpha.txt" was used to define the dictionary. The file can be downloaded here: https://github.com/dwyl/english-words/blob/master/words_alpha.txt .

```python
def dictionary_list_from_file(file):
    file_doc = open(file,"rt")
    dictionary_list = file_doc.read().split("\n")
    return dictionary_list
```

Now we need a function that randomly selects a word from the dictionary to use in the game. Once the word has been selected, it can't be selected from the list again.

```python
def new_word(word_list, unused_word_list=[]):
    import random
    unused_word = False
    while unused_word != True:
        random_num = random.randint(0,len(word_list))
        word = word_list[random_num]
        if word not in unused_word_list:
            unused_word_list.append(word)
            unused_word = True
    return word
```

Let's define functions for creating the guess_string, asking for the user's guess and making sure it's valid, a fill function to update the guess_string, and finally a "play_again" function.

```python
def guess_string(word):
    guess = list(word).copy()
    for i in range(0,len(guess)):
        guess[i] = " _"
    guess_string = ""
    guess_string = guess_string.join(guess) 
    # Creates a string with blank underscores, separated by a space to help the player see how many letters there are in the word. 
    return guess_string.lstrip()

def user_guess(tracking_list):
    guess = ""
    while guess == "":
        guess = str(input("Guess a letter: "))
        if len(guess) != 1: # Player enters too many characters.
            print("You must guess one letter at a time. Try again")
            guess= ""
            continue
        elif guess.isalpha() != True: # Player enters a character that isn't in the English alphabet.
            print("You must guess a single letter from the English alphabet. Try again.")
            guess = ""
            continue
        elif guess in tracking_list:
            print("You've already guessed that letter. Try again.")
            guess = ""
            continue
        else:
            return guess.lower()

def fill_guess_string(guess,string,answer):
    # Updates your guess string to add your guessed letter if it appears in the answer word.
    answer_list = list(answer)
    for i in range(0,len(answer)):
        if answer_list[i] == guess:
            string_list = string.split()
            string_list[i] = guess
            string = ' '.join(string_list)
    return string

def play_again():
    play_question = ""
    while play_question == "":
        play_question = str(input("Would you like to play again? Valid responses are Yes, Y, No, or N :  ")).lower()
        if play_question == 'Yes'.lower() or play_question == 'Y'.lower(): # Sets the game session while loop's condition to stay True
            return True
        elif play_question == 'No'.lower() or play_question == 'N'.lower(): # Sets the game session while loop's condition to False
            return False
        else:
            print("Invalid response. Please enter a valid response.")
            play_question=""
            continue
```

```python
word_list = dictionary_list_from_file("words_alpha.txt")
unused_word_list = []
guess_counter = 7
guess_game = True # Starts the game session
Wins = 0
Loses = 0

while guess_game:
    
    # Initializes each new round
    answer = new_word(word_list,unused_word_list) # Picks word, keeps old word removed from list so next word never repeats
    string = guess_string(answer)
    tracking_list = [] # Creates blank tracking list for each round
    guessing_sequence = True
    
    # A guessing round
    while guessing_sequence:
        print("\n")
        guess = user_guess(tracking_list)
        tracking_list.append(guess)
        print(tracking_list)
        string = fill_guess_string(guess,string,answer)
        
    # Checking win conditions
        answer_list = list(answer)

        # Checks to see if you guessed correctly to determine the appropriate dialogue for the user.
        correct_guess = False
        for i in range(0,len(answer)):
            if answer_list[i] == guess:
                correct_guess = True

        if correct_guess: # If you guessed correctly but haven't gotten the entire word yet
            if answer_list == list(''.join(string.split())):
                Wins += 1
                print(f"\nCongratulations! You discovered the word, {answer}!\nThe letters you guessed this round were {tracking_list}.\nYou had {guess_counter} guesses left.\nYour Win-Lose totals for this session are {Wins}-{Loses}.")
            
                # Once you've reached and end result (win or lose), ask player if they'd like to play again.
                check_play = play_again()
                if check_play == True:
                    guessing_sequence = False # Ends the inner while loop, but starts a new round with a new word.
                elif check_play == False:
                    guessing_sequence = False
                    guess_game = False # Ends both while loops, ending the program.
            else:
                print(f"\nYou guessed a letter that is in the answer!\nHere is an updated version of your word with the guessed letter filled in: {string}.\nThe letters you've guessed so far are: {tracking_list}")
                pass
        
        elif not correct_guess:
        # Take away 1 guess if incorrect
            guess_counter -= 1
            if (guess_counter>0):
                print(f"\nThat letter is not found in the word!\nThe number of guesses you have left are:  {guess_counter}.\nThe letters you've guessed so far are: {tracking_list}.\nThe word now looks like: {string}")
                pass
            elif guess_counter == 0:
                Loses += 1
                print(f"\nYou have run out of guesses and lost the round!\nThe word was: {answer}.\nThe letters you guessed were: {tracking_list}\nYour win-lose totals are now: {Wins}-{Loses}.")

                # Once you've reached and end result (win or lose), ask player if they'd like to play again.
                check_play = play_again()
                if check_play == True:
                    guessing_sequence = False # Ends the inner while loop, but starts a new round with a new word.
                elif check_play == False:
                    break # Ends the program.
```
